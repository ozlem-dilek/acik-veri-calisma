```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("ilce-mahalle-bazli-sulama-suyu-kaynagi-veri.csv")
```

```python
pd.options.display.max_rows = None
```

```python
data.info()
```

```python
data.tail()
```

```python
data.drop(columns="_id", inplace=True)
```

```python
data.drop(index=772, inplace=True)
data.drop(index=676, inplace=True)
data.drop(index=644, inplace=True)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data["veri_sayisi"] = data["tarimsal_sulama_suyu_kaynagi"].str.count(",") + 1
```

```python
data["veri_sayisi"].max()
```

```python
data[["tarimsal_sulama_suyu_kaynagi1", "tarimsal_sulama_suyu_kaynagi2", "tarimsal_sulama_suyu_kaynagi3"]] = data["tarimsal_sulama_suyu_kaynagi"].astype(str).str.split(",", expand=True)
```

```python
data.drop(columns="tarimsal_sulama_suyu_kaynagi", inplace=True)
```

```python
data.drop(columns="veri_sayisi", inplace=True)
```

```python
for index, row in data.iterrows():
    for col in data.columns:
        if pd.isnull(row[col]):
            data.loc[index, col] = "YOK"
```
