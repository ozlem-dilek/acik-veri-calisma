```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("kirsal-ve-kirsal-olmayan-tum-mahalleler-veri.csv")
```

```python
data.info()
```

```python
data.head(100)
```

```python
data.drop(columns="_id", inplace=True)
```

```python
data.rename(columns={"Ilce" : "ilce"}, inplace=True)
```

```python
data['ilce'] = data['ilce'].str.replace('Þ', 'Ş')
data['ilce'] = data['ilce'].str.replace('Ð', 'Ğ')
data['mahalle'] = data['mahalle'].str.replace('Þ', 'Ş')
data['mahalle'] = data['mahalle'].str.replace('Ð', 'Ğ')
```

```python
data.head(500)
```

```python
data['mahalle_turu'] = data['mahalle_turu'].str.replace('KIRSALMAHALLE', 'KIRSAL MAHALLE')
```
