```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("selcuklu_emekli_lokalleri-temizlenmemis-veri.csv", sep=";")
```

```python
data.info()
```

```python
data = data.drop(columns="sunulan_hizmet_1", axis=0)
data = data.drop(columns="sunulan_hizmet_2", axis=0)
data = data.drop(columns="sunulan_hizmet_3", axis=0)
data = data.drop(columns="sunulan_hizmet_4", axis=0)
data = data.drop(columns="id", axis=0)
```

```python
data[["aylik_yararlanici_sayisi", "yillik_yararlanici_sayisi"]] = data["aylık/yıllık_yaralanıcı_sayısı"].astype(str).str.split("/", expand=True)
```

```python
data = data.drop(columns="aylık/yıllık_yaralanıcı_sayısı", axis=0)
```

```python
data["enlem"] = data["enlem"].astype(str).str.replace(",",".")
data["boylam"] = data["boylam"].astype(str).str.replace(",",".")
```

```python
data.dropna(inplace=True)
```

```python
data["yillik_yararlanici_sayisi"] = data["yillik_yararlanici_sayisi"].astype(str).str.replace(".","")
```

```python
data
```

```python
data.rename(columns={"emekli_lokali_adı" : "emekli_lokali_adi"}, inplace=True)
```

```python
data["emekli_lokali_adi"] = data["emekli_lokali_adi"].astype(str).str.title()
```

```python
data["emekli_lokali_adi"] = data["emekli_lokali_adi"].astype(str).str.replace("i̇","i")
```

```python
data["icecek_hizmeti"] = "var"
data["gazete_hizmeti"] = "var"
data["televizyon_hizmeti"] = "var"
```
