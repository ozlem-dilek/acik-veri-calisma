```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("sehir-tiyatrosu-2019.csv")
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(columns="_id", axis=0, inplace=True)
data.drop(columns="gun", axis=0, inplace=True)
```

```python
data["tarih"] = data["tarih"].astype(str).str.replace("T00:00:00","")
```

```python
data["tarih"] = pd.to_datetime(data["tarih"], format="%Y-%m-%d")
```

```python
data["saat"] = data["saat"].astype(str).str.replace(".", ":")
```

```python
data["saat"] = data["saat"].astype(str).str.replace("\x96", " - ")
```

```python
data["saat"][32]
```

```python
data.head(30)
```

```python
data[["seans_1", "seans_2"]] = data['saat'].astype(str).str.split(" - ", expand=True)
```

```python
data.head(30)
```

```python
for index, row in data.iterrows():
    if row['seans_2'] != None:
        new_row = {'tarih' : row['tarih'],
                    'saat' : row['seans_2'],
                    'oyun' : row['oyun'],
                    'doluluk_orani' : row['doluluk_orani'],
                    'izleyici_sayisi' : (row['izleyici_sayisi']/2)
                    }
        data = data.append(new_row, ignore_index=True)
```

```python
data
```

```python
data.drop(columns="seans_1", axis=0, inplace=True)
data.drop(columns="seans_2", axis=0, inplace=True)
```

```python
for i in range(0,len(data)):
    if len(data['saat'][i]) > 5:
        data['saat'][i] = data['saat'][i][:6]
```

```python
for i in range(0,len(data)):      
    if data['saat'][i] == '13:3':
            data['saat'][i] = '13:30'
```

```python
data = data.sort_values(by='tarih')
```
