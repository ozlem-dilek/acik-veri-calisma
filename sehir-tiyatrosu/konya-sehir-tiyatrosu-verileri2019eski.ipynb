```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("sehir-tiyatrosu-2019.csv")
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(columns="_id", axis=0, inplace=True)
data.drop(columns="gun", axis=0, inplace=True)
```

```python
data.head()
```

```python
data["tarih"] = data["tarih"].astype(str).str.replace("T00:00:00","")
```

```python
data["tarih"] = pd.to_datetime(data["tarih"], format="%Y-%m-%d")
```

```python
data
```

```python
data["saat"] = data["saat"].astype(str).str.replace(".", ":")
```

```python
data.head()
```

```python
data
```

```python
for index in range(len(data)):
    if len(data['saat'][index])>5:

        if '-' in data['saat'][index]:
            saatler = data['saat'][index].split('-')
        else:
            saatler = data['saat'][index].split('   ')
        
        for saat in saatler:
            new_row = {'tarih' : data['tarih'][index],
                    'saat' : saat,
                    'oyun' : data['oyun'][index],
                    'doluluk_orani' : data['doluluk_orani'][index],
                    'izleyici_sayisi' : (data['izleyici_sayisi'][index]/2)
                    }
            data = data.append(new_row, ignore_index=True)
```

```python
data
```

```python
for index in range(len(data)):
    if len(data['saat'][index])>5:
        data.drop(index=index, inplace=True, axis=1)
```
