```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("sehir-tiyatrosu-2018.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(columns="_id", axis=0, inplace=True)
data.drop(columns="gun", axis=0, inplace=True)
```

```python
data.head()
```

```python
data["tarih"][0]
```

```python
data["tarih"] = data["tarih"].astype(str).str.replace(".","-")
```

```python
data[["gun", "ay", "yıl"]] = data["tarih"].astype(str).str.split("-", expand=True)
```

```python
data
```

```python
data["tarih"] = data["ay"]+"-"+data["gun"]+ "-" +data["yıl"]
```

```python
data.drop(columns="gun", axis=0, inplace=True)
data.drop(columns="ay", axis=0, inplace=True)
data.drop(columns="yıl", axis=0, inplace=True)
```

```python
data["tarih"][17]=data["tarih"][17].replace(" ", "")
```

```python
data["tarih"] = pd.to_datetime(data["tarih"], format="%m-%d-%Y")
```

```python
data["saat"] = pd.to_timedelta(data["saat"], unit='h')
```

```python
data["tarih"] = data["tarih"] + data["saat"]
```

```python
data.drop(columns="saat", axis=0, inplace=True)
```

```python
data.rename(columns={"tarih" : "tarih_saat"}, inplace=True)
data.rename(columns={"oyun" : "oyun_adi"}, inplace=True)
```
